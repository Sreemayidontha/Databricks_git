```python
# Doctors - Data and columns
data=[
        [10, "Dr. Smith", "Cardiology"],
        [11, "Dr. Lee", "Neurology"],
        [12, "Dr. Patel", "Cardiology"]
    ]
 
columns=["doctor_id", "doctor_name", "specialty"]
 
df_doctors=spark.createDataFrame(data,columns) 
# Visits - Data and columns
data=[
        [1001, 1, 10, "2024-01-05", 200],
        [1002, 1, 11, "2024-02-10", 300],
        [1003, 2, 10, "2024-01-15", 150],
        [1004, 2, 12, "2024-02-20", 400],
        [1005, 3, 12, "2024-02-01", 250],
        [1006, 4, 11, "2024-01-25", 350]
    ]
columns=["visit_id", "patient_id", "doctor_id", "visit_date", "cost"]
 
df_visits=spark.createDataFrame(data,columns) 
```

```python
display(df_doctors)
```

```python
display(df_visits)
```

```python
df_join = df_doctors.join(df_visits, on=df_doctors.doctor_id == df_visits.doctor_id, how="inner")
```

```python
display(df_join)
```

```python
df1=df_join
```

```python
df1 = df_join.groupBy("doctor_id", "doctor_name", "specialty").agg(sum("cost").alias("total_earnings"))
```
